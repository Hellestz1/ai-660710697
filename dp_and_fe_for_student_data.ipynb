```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts' , 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวะ', 'Engineering', 'วิศวฯ' , 'วิศวกรรม']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
std['คณะ'] = std['คณะ'].apply(standardize_faculty)
```

```python
def standardize_gender(gender):
    if gender in ['Male', 'M' , 'ช' ]:
        return 'ชาย'
    elif gender in ['Female', 'F' , 'ญ' ]:
        return 'หญิง'
    else:
        return gender
std['เพศ'] = std['เพศ'].apply(standardize_gender)
```

```python
std.loc[(std['GPA'] < 1) | (std['GPA'] > 4), 'GPA'] = np.nan

Q1 = std['อายุ'].quantile(0.25)
Q3 = std['อายุ'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print(upper_bound)
std.loc[(std['อายุ'] < lower_bound) | (std['อายุ'] > upper_bound) ,'อายุ'] = np.nan
```

```python
mean_gpa = std['GPA'].mean()
std['GPA'] = std['GPA'].fillna(mean_gpa.round(2))
```

```python
std
```

```python
std.isnull().sum()
```

```python
std['ชื่อ'].fillna('ไม่ระบุ', inplace=True)
std['อายุ'].fillna(std['อายุ'].mode()[0], inplace=True)
std['เพศ'].fillna(std['เพศ'].mode()[0], inplace=True)
std['คณะ'].fillna(std['คณะ'].mode()[0], inplace=True)
std['ส่วนสูง_cm'].fillna(std['ส่วนสูง_cm'].mode()[0], inplace=True)
std['น้ำหนัก_kg'].fillna(std['น้ำหนัก_kg'].mode()[0], inplace=True)
std['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(std['ชั่วโมงเรียน_ต่อสัปดาห์'].mode()[0], inplace=True)
```

```python

# ทำ One-Hot Encoding
data_encoded = pd.get_dummies(std, columns=['คณะ'])
```

```python
# สร้าง BMI
std['BMI'] = std['น้ำหนัก_kg'] / (std['ส่วนสูง_cm']/100)**2
std['สุขภาพ'] = pd.cut(std['BMI'], bins=[0, 18.5, 25, 30, 100],labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```
